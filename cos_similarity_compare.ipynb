```python
import pandas as pd
import matplotlib.pyplot as plt

chatgpt = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/COMP0087-NLP/chatgpt_gpt_cos_similarity.csv")
human = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/COMP0087-NLP/huamn_gpt_cos_similarity.csv")
```

```python
chatgpt_cos = []
for i in range(len(chatgpt['y_score'])):
  chatgpt_cos.append(chatgpt['y_score'][i])
print(chatgpt_cos)
```

```python
human_cos = []
for i in range(len(human['y_score'])):
  human_cos.append(human['y_score'][i])
print(human_cos)
```

```python
plt.hist(chatgpt_cos, bins=150, alpha=1, label='chatgpt')  
plt.hist(human_cos, bins=150, alpha=0.65, label='human')  
plt.xlabel('cosine similarity')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.xlim(2, 7)
plt.legend()
plt.show()
```
